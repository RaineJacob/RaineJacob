```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Reading Data


```python
data=pd.read_csv("D:\\Datasets\\BhartiAirtel.csv")
data.head()
data.tail()
```

Taking Info Out of Dataset

```python
data.info()
```

Describing Dataset

```python
data.shape
```

```python
data.describe()
```

Graph Of Opening Price

```python
data['Open Price'].plot()
plt.title("Bharti Airtel opening price 2011-2021")
plt.show
```

Graph Of Closing Price

```python
data['Close Price'].plot()
plt.title("Bharti Airtel closing price 2011-2021")
plt.show
```

Average Of High and Low Price

```python
AvgHigh = data.describe()['High Price']['mean']
Avglow = data.describe()['Low Price']['mean']

print("Average High Bharti Airtel price is :",AvgHigh,
      "\nAverage Low Bharti Airtel price is :",Avglow,
    )
```

All Prices Together

```python
data['Open Price'].plot(label='Bharti Airtel Open Price')
data['Close Price'].plot(label='Bharti Airtel Close Price')
data['High Price'].plot(label = 'Bharti Airtel High Price')
data['Low Price'].plot(label='Bharti Airtel Low Price ')
plt.legend()
plt.title('Bharti Airtel Prices')
plt.ylabel("Stock Price")
plt.show()
```

Histogram of Weighted Average Price

```python
plt.hist(data.WAP,np.arange(100,900,100))
```

Graph of WAP

```python
data['WAP'].plot(figsize=(17,5))
plt.title("Bharti Airtel VOlume 2011-2021")
plt.show
```

Spike Data

```python
spike=data.iloc[[data['WAP'].argmax()]]
spike
```

```python
spike = data.iloc[250:300]
spike['Open Price'].plot()
```
